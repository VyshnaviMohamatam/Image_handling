```python
video=cv2.VideoCapture(r"C:\Users\DELL\Downloads\3173312-uhd_3840_2160_30fps.mp4")
```

```python
while True:
    succ,img=video.read()
    if succ==False:
       break
    b,g,r=cv2.split(img)
    z=np.zeros(b.shape,dtype=np.uint8)

    cv2.imshow("blue_chanel",cv2.merge([b,z,z]))
    cv2.imshow("green_chanel",cv2.merge([z,g,z]))
    cv2.imshow("red_chanel",cv2.merge([z,z,r]))
    cv2.imshow("video_gray",img1)
    if cv2.waitKey(1) & 255 == ord("A"):
        break
cv2.destroyAllWindows()
```

```python
l=[]
for y in os.listdir(r"C:\Users\DELL\Downloads\cat-dog-elephant"):
    if y.endswith(".mp4"):
        l.append(y)
    
    
l
```

```python
l[0].split(".")[0]
```

```python
for video in l:
    vid=cv2.VideoCapture(r"C:\Users\DELL\Downloads\cat-dog-elephant\{}".format(video))
    c=0
    while True:
        succ,img=vid.read()
        if succ==False:
            break
        cv2.imwrite(r"C:\Users\DELL\Downloads\cat-dog-elephant\{}\{}{}{}.jpg".format(video.split(".")[0],video.split(".")[0],video.split(".")[0],c),img)
        c=c+1
        cv2.imshow("vid",img)
        if cv2.waitKey(1) & 255 == ord("q"):
            break
    cv2.destroyAllWindows()
        
```
