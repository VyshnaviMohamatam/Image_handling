```python
import os
import cv2
```

```python
folders=os.listdir(r"C:\Users\DELL\Downloads\archive\flower_photos")
folders
```

```python
features=[]
class_labels=[]
for folder in folders:
    for images in os.listdir(r"C:\Users\DELL\Downloads\archive\flower_photos\{}".format(folder)):# listdir returns the files in a folder#
        img=cv2.imread(r"C:\Users\DELL\Downloads\archive\flower_photos\{}\{}".format(folder,images),0)#imread read the image for particular folder#
        img=cv2.resize(img,(50,50))#resize size of the image dimensions#
        img=img.flatten()# flatten is used to convert nd array into 1d array# 
        features.append(img)
        class_labels.append(folder)
```

```python
import pandas as pd
```

```python
final_data=pd.DataFrame(features)
final_data
```
